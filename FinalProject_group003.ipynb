## Background & Prior Work

The 2019-20 NBA Bubble happened as a result of the COVID-19 pandemic and it took place in Disneyworld in Orlando, Florida. When the country went into quarantine the NBA suspended play because sporting events were a major concern when it came to spreading the virus and especially indoor arenas. As a result, the NBA created a “bubble” where only necessary staff members and players of a select number of basketball teams were allowed to attend. They were required to quarantine in their hotel rooms and were not allowed to leave the bubble or else they would have to quarantine again. One of the most significant impacts of the bubble was the absence of fans at the games. The fans often play a major part in the atmosphere of a basketball game. The NBA has never had an entire playoffs, where the best teams compete for a championship, without any fans in attendance.
   
The 3-point shot has seen a dramatic increase in recent years and the increased usage of this has changed the way teams, players, and coaches approach the sport of basketball. Thus when comparing different seasons regarding performance it makes sense to compare seasons that are closer in time.

## Data Analysis Tests & Results

The tests that we chose to use on the data to determine whether the bubble affected the NBA players' performance were 2 sample t-test for FG% and 3P% statistics. This is because we are using two independent distributions of precovid and postcovid data that are normally distributed. We used the Mann-Whitney test for FT% statistic because the data for FT% was not normally distributed. We did not test +/- data because this data by the nature of how the statistic was calculated always had a mean of zero and was therefore not a suitable statistic to use for comparison between different datasets.

```python
pre_covid_FG = pre_covid['FG%']
during_covid_FG = during_covid['FG%']
stats.ttest_ind(pre_covid_FG, during_covid_FG, equal_var=False)
```

The 2 sample t-test for FG% has a p-value of 0.193 which is greater than alpha=0.05 so we fail to reject our null hypothesis that there is no difference between pre-covid field goal percentages and during covid field goal percentages. There is insufficient statistically significant evicence to suggest that playing during the bubble negatively impacted field goal performance.

```python
pre_covid_3P = pd.concat([players1415_3P['3P%'], players1516_3P['3P%'], players1617_3P['3P%'], 
                          players1718_3P['3P%'], players1819_3P['3P%']], axis=0)
during_covid_3P = players1920_3P['3P%']
stats.ttest_ind(pre_covid_3P, during_covid_3P, equal_var=False)
```

The 2 sample t-test for 3P% has a p-value of 0.114 which is greater than alpha=0.05 so we fail to reject our null hypothesis that there is no difference between pre-covid 3 point goal percentages and during covid 3 point goal percentages. There is insufficient statistically significant evicence to suggest that playing during the bubble negatively impacted 3 point performance.

```python
pre_covid_FT = pd.concat([players1415_FT['FT%'], players1516_FT['FT%'], players1617_FT['FT%'], 
                          players1718_FT['FT%'], players1819_FT['FT%']], axis=0)
during_covid_FT = players1920_FT['FT%']
stats.mannwhitneyu(pre_covid_FT, during_covid_FT, alternative='two-sided')
```

The Mann-Whitney test for FT% has a p-value of 0.454 which is greater than alpha=0.05 so we fail to reject our null hypothesis that there is no difference between pre-covid free throw percentages and during covid free throw percentages. There is insufficient statitically significant evidence to suggest that plyaing during the bubble negatively impacted free throw performance.
